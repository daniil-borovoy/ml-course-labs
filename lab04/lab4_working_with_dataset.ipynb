### Обработка пропусков данных

Один из способов простых способов решения этой проблемы - просто игнорировать или удалять строки, в которых отсутствуют данные, выбрасывая их из нашего анализа. Однако этот метод может быть плох из-за потери информации.

Еще один способ — это заполнение пропусков, где мы заменяем отсутствующее значение каким-либо образом. Базовые реализации просто заменят все отсутствующие значения средним, медианным, либо же константой.



```python
missing_counts = features.isnull().sum()
missing_counts.sort_values(ascending=False).head(20)
```

```python
df['PoolQC'] = df['PoolQC'].fillna('NA')
df['FireplaceQu'] = df['FireplaceQu'].fillna('NA')

# Заменяем пропуски в категориальных признаках модой
categorical_columns = ['MiscFeature', 'Alley', 'Fence',
                       'GarageType', 'GarageQual', 'GarageCond', 'GarageFinish',
                       'BsmtFinType2', 'BsmtExposure', 'BsmtCond', 'BsmtFinType1', 'BsmtQual',
                       'MasVnrType', 'Electrical']

for col in categorical_columns:
    df[col].fillna(df[col].mode()[0], inplace=True)

# Заменяем пропуски в числовых признаках средним значением
numeric_columns = ['LotFrontage', 'GarageYrBlt', 'MasVnrArea']

for col in numeric_columns:
    df[col].fillna(df[col].mean(), inplace=True)

# Выводим обновленный DataFrame
print(df)
```

### Обнаружение выбросов

```python
import matplotlib.pyplot as plt
import seaborn as sns

print(df.describe(include=['number']).loc[['min','max','mean']].T.sort_values('max'))
outlines_data = df.describe(include=['number'])


# Добавить графики и найти выбросы


fig, axes = plt.subplots(6, 6, figsize=(18, 18))

for i, feature in enumerate(outlines_data):
    row, col = i // 6, i % 6
    ax = axes[row, col]
    sns.histplot(data=outlines_data, x=feature, kde=True, ax=ax)
    ax.set_title(f'{feature}')

plt.tight_layout()
plt.show()


for column in df.select_dtypes(include=['number']).columns:
  Q1 = df[column].quantile(0.25)
  Q3 = df[column].quantile(0.75)
  IQR = Q3 - Q1
  is_outlier = ~df[column].between(Q1 - 1.5 * IQR, Q3 + 1.5 * IQR)

  if is_outlier.any():
    sns.boxplot(df[column])
    plt.title(f'Boxplot of {column}')
    plt.show()
```

### Кодирование данных

```python
# Выберите только категориальные признаки (object dtype)
categorical_features = df.select_dtypes(include=['object'])

# One-Hot Encoding
data_encoded = pd.get_dummies(df, columns=categorical_features.columns)
```

### Зависимости в данных

```python
# !!! заюзать data_encoded

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

correlation_matrix = data_encoded.corr()

kot = correlation_matrix[correlation_matrix >= 0.8]
kot = kot[kot < 1]
print([item for item in kot if item is not None])
plt.figure(figsize=(12,8))
sns.heatmap(kot, cmap="coolwarm")

columns = [item for item in kot if item is not None]

data_encoded.drop(columns, axis=1)

correlation_matrix = data_encoded.corr()
kot = correlation_matrix[correlation_matrix >= 0.8]
kot = kot[kot < 1]

plt.figure(figsize=(8, 6))
sns.heatmap(kot, annot=True, cmap='coolwarm', fmt=".2f")
plt.show()
```
